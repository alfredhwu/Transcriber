```python
""" Belle-whisper """
from transformers import pipeline

mm = "BWLv3zhp"
audiofile = "4月14日 上午10点19分.m4a"

transcriber = pipeline(
  "automatic-speech-recognition", 
  device="mps",
  model="BELLE-2/Belle-whisper-large-v3-zh-punct",
  return_timestamps=True
)

transcriber.model.config.forced_decoder_ids = (
  transcriber.tokenizer.get_decoder_prompt_ids(
    language="zh", 
    task="transcribe"
  )
)

transcription = transcriber(f"recordings/{ audiofile }") 

with open(f"transcriptions/{ mm }-{ audiofile }.txt", "w") as writer:
    writer.write(transcription["text"])
```
